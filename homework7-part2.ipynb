```python
#We want: titles, subhead, article URL, whether it's premium or not, byline, article type, image URL.

import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.lemonde.fr/en/'
response = requests.get(url)

soups = BeautifulSoup(response.text, 'html.parser')
soups
```

```python
rows = []
for soup in soups:
    try:
        title = soup.select('.article__title-label, .article__title').get_text(strip=True)
    except:
        title = 'Not found'
    try:
        title_url = soup.find('a').get('href', None)
    except:
        title_url = 'Not found'
    try:
        subhead = soup.find('p', attrs={'class':'article__desc'}).get_text(strip=True)
    except:
        subhead = 'Not found'
    try:
        article_type = soup.select_one('a.ds-breadcrumb__link')['aria-label'].get_text(strip=True)
    except:
        article_type = 'Not found'
    try:
        byline = soup.find('span', attrs={'class':'meta__author'}).get_text(strip=True)
    except:
        byline = 'Not found'
    try:
        image_url = soup.select_one('.article__media img').get('src', None)
    except:
        image_url = 'Not found'
    try:
        premium = soup.find(class_='ds-article-status__text').get_text(strip=True)
    except:
        premium = 'Not found'

    row =  {}

    row['title'] = title
    row['title_url'] = title_url
    row['subhead'] = subhead
    row['article_type'] = article_type
    row['byline'] = byline
    row['image_url'] = image_url
    row['premium'] = premium

    rows.append(row)
rows
    
```

```python
import pandas as pd
df = pd.DataFrame(rows)
df
df.to_csv('homework7_part2.csv')
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time

options = Options()

# ヘッドレスは失敗しやすいので最初はOFF
# options.add_argument("--headless")

# Bot対策を少し回避
options.add_argument("--disable-blink-features=AutomationControlled")

driver = webdriver.Chrome(options=options)

driver.get(article_url)

# 読み込み待ち
time.sleep(10)

# HTML取得
html = driver.page_source
print(html[:1000])  # 確認

driver.quit()
```

```python
soup_detail = BeautifulSoup(html, 'html.parser')
soup_detail
```
