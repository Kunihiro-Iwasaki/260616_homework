```python
import requests
from bs4 import BeautifulSoup

url = 'https://travel-id-documents.az.gov/authorized-third-party-driver-license-locations'

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36" 
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
```

```python
locations = soup.find_all('td')
locations
```

```python
company = locations[8].get_text(strip=True)
print(company)
url = locations[8].find('a').get('href', None)
print(url)
adress = locations[9].get_text(strip=True)
print(adress)
tel = locations[10].get_text(strip=True)
print(tel)
hours = locations[11].get_text(strip=True)
print(hours)
```

```python
rows = []
for i in range(4, len(locations), 4):
    try:
        company = locations[i].get_text(strip=True)
    except:
        company = 'Not found'
    try:
        url = locations[i].find('a').get('href', None)
    except:
        url = 'Not found'
    try:
        adress = locations[i+1].get_text(strip=True)
    except:
        adress = 'Not found'
    try:
        tel = locations[i+2].get_text(strip=True)
    except:
        tel = 'Not found'
    try:
        hours = locations[i+3].get_text(strip=True)
    except:
        hours = 'Not found'

    row = {}
    row['company'] = company
    row['url'] = url
    row['adress'] = adress
    row['tel'] = tel
    row['hours'] = hours

    rows.append(row)
rows
```

```python
import pandas as pd
df = pd.DataFrame(rows)
print(df)
df.to_csv('homework7_part4.csv', index=False)
```
