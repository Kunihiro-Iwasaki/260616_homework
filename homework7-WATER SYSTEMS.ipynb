```python
import requests
from bs4 import BeautifulSoup
url = 'https://dww.des.sc.gov/DWW/JSP/SearchDispatch?number=&name=&county=All&WaterSystemType=All&SourceWaterType=All&PointOfContactType=None&SampleType=null&begin_date=6%2F19%2F2024&end_date=6%2F19%2F2026&action=Search+For+Water+Systems'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
items = soup.find_all('td')
items[10:15]
```

```python
try:
    ws_number = items[10].get_text(strip=True)
    print(ws_number)
except:
    ws_number = 'None'
try:
    url = items[10].find('a').get('href', None)
    print(url)
except:
    url = 'None'
try:
    ws_name = items[11].get_text(strip=True)
    print(ws_name)
except:
    ws_name = 'None'
try:
    ws_type = items[12].get_text(strip=True)
    print(ws_type)
except:
    ws_type = 'None'
try:
    status = items[13].get_text(strip=True)
    print(status)
except:
    status = 'None'
try:
    principal = items[14].get_text(strip=True)
    print(principal)
except:
    principal = 'None'
try:
    source = items[15].get_text(strip=True)
    print(source)
except:
    source = 'None'
    
```

```python
rows = []
for j in range(10, len(items), 6):
    
    try:
        ws_number = items[j].get_text(strip=True)
        print(ws_number)
    except:
        ws_number = 'Not found'
    
    try:
        url = items[j].find('a').get('href', None)
        print(url)
    except:
        url = 'Not found'
    
    try:
        ws_name = items[j+1].get_text(strip=True)
        print(ws_name)
    except:
        ws_name = 'Not found'
    
    try:
        ws_type = items[j+2].get_text(strip=True)
        print(ws_type)
    except:
        ws_type = 'Not found'
    
    try:
        status = items[j+3].get_text(strip=True)
        print(status)
    except:
        status = 'Not found'
    
    try:
        principal = items[j+4].get_text(strip=True)
        print(principal)
    except:
        principal = 'Not found'
    
    try:
        source = items[j+5].get_text(strip=True)
        print(source)
    except:
        source = 'Not found'

    row = {}
    row['ws_number'] = ws_number
    row['url'] = url
    row['ws_name'] = ws_name
    row['ws_type'] = ws_type
    row['status'] = status
    row['principal'] = principal
    row['source'] = source

    rows.append(row)
rows
```

```python
import pandas as pd
df = pd.DataFrame(rows)
df.tail()
```

```python
df = df.drop(1392)
df
```

```python
df.to_csv('homework7-watersystems.csv', index=False)
```
